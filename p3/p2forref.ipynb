### 2nd question : About cholesterol, I would like to know if trans-fat has an effect compared to traditional fat

```python
# First of all, I create a dataframe only with interresing columns, 
# and also rows by removing products without information about transfat or cholesterol
transfat = 'trans-fat_100g'
cholesterol = 'cholesterol_100g'
df3_fat_transfat = df[[fat,transfat, cholesterol]]
df3_fat_transfat = df3_fat_transfat.dropna (how = 'any')
df3_fat_transfat.describe()
```

```python
# First, we can clean it ... some products have more than 100g or transfat for 100g of product !
df3_fat_transfat = df3_fat_transfat[df3_fat_transfat[transfat] < 100]
df3_fat_transfat.describe()
```

```python
# I create a new column which is the part of Trans_fat in the total Fat 
# meaning TransFat / Fat
TransfatPerFat = 'transfat_per_fat'
df3_fat_transfat[TransfatPerFat] = df3_fat_transfat[transfat] / df3_fat_transfat[fat]
df3_fat_transfat[TransfatPerFat].describe()
```

```python
# ohoh. We have values to clean. The part of Trans fat cannot be > 1
df3_fat_transfat = df3_fat_transfat[df3_fat_transfat[TransfatPerFat] < 1]
df3_fat_transfat[TransfatPerFat].describe()
```

```python
sns.regplot(df3_fat_transfat[TransfatPerFat], df3_fat_transfat[cholesterol], fit_reg = False, x_bins=np.linspace(0, 1.0, num=20))
plt.ylim(-.05, .2)
plt.xlim(-0.05, 1)
plt.show()

# we can see here that cholesterol doesn't depends on proportion of transfat ...
#the more  trandfat we have, bigger is the  dispersoin ..
```

# D. Time-series analysis

```python
# The dataframe used here is the df3 (cleaned by my way)

#Addtional cleaning to ensure both dates are filled
df3 = df3.dropna(subset  = ['last_modified_datetime', 'created_datetime' ], how = 'any')

#Let's compute difference between the two times

Delta = df3['last_modified_datetime'] - df3['created_datetime'] 
```

```python
# before computing the mean, let's see the shape of Delta (in seconds here)
DeltaSec = Delta.apply(lambda x: x.total_seconds())
pc = DeltaSec.values
#plt.ylim((0,500))
plt.hist(pc)
plt.show()
```

```python
# Ok .. It will be difficult to interpret the mean, but we can compute it by he way
# We can deduce the mean by aggregating the result
Delta.mean()
```

```python
# Now, let's find  the first and last creations dates

MinTime = df3['created_datetime'].min()
MaxTime = df3['created_datetime'].max()
(MinTime, MaxTime)
```

```python
# I would like to use created_datetime as an index. But for that, it must be unique.
# Let's create  a dataframe with :
# - created_datetime as an index
# - one column telling how many products are created for corresponding created_datetime
d1 = df3['created_datetime']
d1 = d1.groupby(d1).count().to_frame()

# Let's see a litle bit the result.
d1.head()
```

```python
# I sort values to see if the count() function wrked as expected
d1.sort_values(by = 'created_datetime', ascending=False).head(5)
```

# E. Build a database

```python
#First, let's connect to a new database
import sqlite3
db = sqlite3.connect('database.sqlite')
```

## First Table : ALL the used ingredients 

```python
# My first table will be ingredients (Key, Text).
# By taking ingredients used before, I create a new DataFrame : the index will be used as a PRIMARY KEY for the DataBase
dfIngredients = pd.DataFrame({'Ingredients':list(t5.unique())})
dfIngredients.head()
```

```python
query = 'CREATE TABLE "ingredients" ( "index" INTEGER PRIMARY KEY, "Ingredients" TEXT )'
cursor = db.cursor()
cursor.execute(query)
dfIngredients.to_sql(name='ingredients', con=db, if_exists='replace', index=True)
```

## Second Table : List of EVERY ingredients of EVERY products

```python
# My second table will be a list of ALL ingredients of ALL products. Primary key is (Product Index x Numero of ingredient)
# The only useful column is the ID of the ingredient (index of "ingredients" table).
```

```python
#First of all, we want to store in the database index of ingredient, not its text

#Let's 'inverse"' the dfIngredients used just above to use it as a map
dfIngredientsInv = dfIngredients.reset_index(level=0).set_index('Ingredients')
dfIngredientsInv.head()
```

```python
# Here is the list of ALL the ingredients of all products.
t5.head()
```

```python
#Replace text of ingredient with ID of the ingredients table
list_ingredients = t5.map(dfIngredientsInv['index'])
list_ingredients.head(15)
```

```python
query = 'CREATE TABLE "list_ingredients" ( "level_0" INTEGER, "level_1" INTEGER, "0" INTEGER, PRIMARY KEY (level_0, level_1)  )'
cursor.execute(query)
list_ingredients.to_sql(name='list_ingredients', con=db, if_exists='replace', index=True)
```

## Third table : list of ALL products (subset actually)

```python
# restrict your data to 1000 entries and 5 columns of your choice
df_subset = df3[[
 'code',
 'product_name', 
 'energy_100g',
 'fat_100g',
 'saturated-fat_100g',
 'trans-fat_100g',
]].head(1000)

#see what it look like ..
df_subset.head()
```

```python
# let's store this product list into a new table (Primary Key is the index of the DataFrame)

query = 'CREATE TABLE "products" ( "index" INTEGER PRIMARY KEY, "code" TEXT, "product_name" TEXT, "energy_100g" REAL, "fat_100g" REAL, "saturated-fat_100g" REAL, "trans-fat_100g" REAL )'
cursor.execute(query)
df_subset.to_sql(name='products', con=db, if_exists='replace', index=True)
```

# F. Finding correlations

```python
# To find correlations between values, first have a look at EVERY couple 

cor = df3.corr().unstack()
lcor = list(cor.items())
lcor
```

```python
# Values are in double (correlation is symetric). 
# Filter this list to avoid doublons and avoid to correlate the same distribution that is always equal to 1

lcorfiltered = list(filter(lambda x:x[0][0] < x[0][1], lcor))
lcorfiltered
```

```python
# We can now sort this list to see interresting correlations 
# We keep absolute value for correlation to see correlated values

# Here are the 10 most correlated couples
list(reversed(sorted(lcorfiltered,key=lambda x:abs(x[1]))))[0:10]

# We can see that nutrition-score from France and Uk are correlated ! good news
# We can see that sodium and salt are very correlated (due to the fact that sodium is the main component of salt)
# And energy/fat are correlated (we all know about that ....)
#...
```

```python
# Let's focus now on nutrition-score (we take france for this exercice)
# See correlation with other data
# I don't take the absolute value for correlation to see if a value tends to lead to an higer nutrition-score or lower nutrition-score

nutrition = 'nutrition-score-fr_100g'
othercat = list(df3._get_numeric_data())
othercat.remove(nutrition)
corrs = [(o, df3[nutrition].corr(df3[o])) for o in othercat]
list(reversed(sorted(corrs,key=lambda x:x[1])))

# We can see here that nutrition-score-uk is very correlated with french score (already described above)
# And then than high nutrition score is mostly linked to fat/energy.
# We can see that the nutrition score doesn't take so much in account the amount of vitamin. 
# The correlation with vitamin/fiber is negative, meaning the more we have vitamins/fiber, the less the nutrition score is.

# Good appetit
```
